```python
# Initialize a list to keep track of columns containing 'X'
columns_with_x = []

# Loop through each column to check for 'X'
for column in df.columns:
    if df[column].dtype == object:  # Ensure we only check object type columns
        if df[column].str.contains('X', na=False).any():  # Check if 'X' is present in the column
            columns_with_x.append(column)

# Replace 'X' with the mode value in the identified columns
for column in columns_with_x:
    mode_value = df[column].mode()[0]  # Calculate the mode (most frequent value) of the column
    df[column] = df[column].replace('X', mode_value)

# Output the result
print(f"Replaced 'X' with mode values in the following columns: {columns_with_x}")
```

```python
df.head(20)
```

```python
# Verify 'X' values are no longer present
print("Checking for 'X' values in the DataFrame:")
x_values_present = df.apply(lambda x: x.astype(str).str.contains('X').any(), axis=0)
print(x_values_present)
```

```python
mode_property_type = df['PropertyType'].mode()[0]

# Replace values containing 'X' with the mode
df['PropertyType'] = df['PropertyType'].apply(lambda x: mode_property_type if 'X' in str(x) else x)
```

```python
# Remove rows where 'MSA' contains 'X'
df = df[~df['MSA'].str.contains('X')]
```

```python
# Remove rows where 'PostalCode' is 'X'
df = df[df['PostalCode'].str.strip().str.upper() != 'X']

# Verify that 'X' has been removed
if 'X' in df['PostalCode'].unique():
    print("There are still 'X' values in the 'PostalCode' column.")
else:
    print("All 'X' values have been successfully removed from the 'PostalCode' column.")
```

```python
mode = df['PropertyState'].mode()[0]

# Replace values containing 'X' with the mode
df['PropertyState'] = df['PropertyState'].apply(lambda x: mode if 'X' in str(x) else x)
```

```python
df['NumBorrowers'] = df['NumBorrowers'].replace('X', pd.NA)

# Convert the column to numeric, forcing any remaining non-numeric values to NaN
df['NumBorrowers'] = pd.to_numeric(df['NumBorrowers'], errors='coerce')

# Calculate the mean value of the column, ignoring NaNs
mean_value = df['NumBorrowers'].mean()

# Replace NaN values with the calculated mean
df['NumBorrowers'].fillna(mean_value, inplace=True)
```
